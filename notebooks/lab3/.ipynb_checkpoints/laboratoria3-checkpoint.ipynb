```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("okienka").getOrCreate()
spark
```

```python
czujnik_temperatury = ((12.5, "2019-01-02 12:00:00"),
(17.6, "2019-01-02 12:00:20"),
(14.6,  "2019-01-02 12:00:30"),
(22.9,  "2019-01-02 12:01:15"),
(17.4,  "2019-01-02 12:01:30"),
(25.8,  "2019-01-02 12:03:25"),
(27.1,  "2019-01-02 12:02:40"),
)

# jeśli potrzebujemy ramki w Sparku to należy ustawić jej schemat
```

```python
from pyspark.sql.functions import to_timestamp
from pyspark.sql.types import StructType, StructField, StringType, DoubleType

schema = StructType([
    StructField("temperatura", DoubleType(), True), # typ liczba, tru określa czy moga byc wartosci brakujace
    StructField("czas", StringType(), True),
])

#pyspark sql bo działamy na ramkach danych
```

```python
SCHEMA = """
temperatura DOUBLE, czas STRING 
""" # DDL
```

```python
df = (spark.createDataFrame(czujnik_temperatury, schema=schema)
      .withColumn("czas", to_timestamp("czas")))
df1 = (spark.createDataFrame(czujnik_temperatury, schema=SCHEMA)
      .withColumn("czas", to_timestamp("czas")))


df.printSchema()
```

```python
df1.printSchema()
```

```python
df.show(3)
df1.show(3)
```

```python
df1.describe().show()
```

```python
df.createOrReplaceTempView("df") # stworz albo zamień, towrzymy widok (view) bo nie zużywa nam pamięci, albo aby nie było dostępu do surowych tabel - chcemy widoki aby udostępnić użytkownikom nie wszystkie dane tylko te które s konieczne - jeśli ktoś coś zepsuje na widoku to widok możemy odtworzyć - bo to nie jest tabela (bezpieczeństwo) 
```

```python
spark.sql("select czas, temperatura from df where temperatura > 21").show(5)
```

```python
# Tumbling window

import pyspark.sql.functions as F

df2 = df.groupBy(F.window("czas","30 seconds")).count()
df2.show(truncate=False)
```
